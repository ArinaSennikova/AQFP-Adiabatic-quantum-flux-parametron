```python
import sys
import os
from pathlib import Path

import numpy as np
import math
from functools import partial
```

```python
path = "C:\josim_projects"
file_in = 'AQFP_MIXER1'  #input .cir file
file_out = file_in #output .dat and .csv files

os.chdir(path)
os.system(f"josim-cli ./{file_in}.cir -V 1")
```

```python
data = np.genfromtxt(f"{file_out}.dat", delimiter=None, skip_header=1)
time = data[:, 0] * 10**9
Iin = data[:, 1]
Ir2 = data[:, 2]
I_L0qA = data[:, 3]
I_L0qB = data[:, 4]
Ibb = data[:, 5]
Ifix = data[:, 6]
Vout = data[:, 7]
print(data.shape) #узнать размер массива
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(6, 1, figsize=(8, 12))
axes[0].plot(time, Ir2, 'b-', linewidth=2)
axes[0].set_ylabel('Ir2',fontsize= 15)
axes[0].grid(True, alpha=0.3)

axes[2].plot(time, Iin, 'r-', linewidth=2,label='Iin')
axes[2].plot(time, Ifix, 'r-', linewidth=2,label='Ifix')
axes[2].legend(fontsize=14)
axes[2].grid(True, alpha=0.3)

axes[1].plot(time, Ibb, 'r-', linewidth=2)
axes[1].set_ylabel('Ibb',fontsize= 15)
axes[1].grid(True, alpha=0.3)

axes[3].plot(time, I_L0qA, 'r-', linewidth=2)
axes[3].set_ylabel('I_L0qA',fontsize= 15)
axes[3].grid(True, alpha=0.3)

axes[4].plot(time, I_L0qB, 'r-', linewidth=2)
axes[4].set_ylabel('I_L0qB', fontsize= 15)
axes[4].grid(True, alpha=0.3)

y_min = min(min(I_L0qA), min(I_L0qB))
y_max = max(max(I_L0qA), max(I_L0qB))
axes[3].set_ylim(y_min, y_max)
axes[4].set_ylim(y_min, y_max)


axes[5].plot(time, Vout, 'r-', linewidth=2)
axes[5].set_xlabel('time, ns', fontsize=15)
axes[5].set_ylabel('Vout', fontsize= 15)
axes[5].grid(True, alpha=0.3)
axes[5].set_ylim(y_min, y_max)

for ax in axes:
    ax.tick_params(axis='both', labelsize=12)

"""
# plt.plot(time, Iin,label='Iin')
#plt.xlabel('time, ns') 
#plt.plot(time, Ir2,label='Ir2')
#plt.legend()
#plt.grid()
#plt.xticks()
#plt.yticks()
#plt.show()
# plt.plot(time, Ibb)
# plt.xlabel('time, ns')
# plt.ylabel('Ibb, mA')
plt.grid()
plt.xticks()
plt.yticks()
plt.show()
plt.plot(time, I_L0qA)
plt.xlabel('time, ns')
plt.ylabel('I_l0qA, mA')
plt.grid()
plt.show()
plt.plot(time, I_L0qB)
plt.xlabel('time, ns')
plt.ylabel('I_l0qB, mA')
plt.grid()
plt.show()
"""
```
